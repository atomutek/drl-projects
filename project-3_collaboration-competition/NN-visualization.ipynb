# Visualization of the networks

```python
from torchviz import make_dot, make_dot_from_trace
from torchviz import make_dot
from torch.autograd import Variable
import torch
```

```python
from model import Actor, Critic
```

```python
action_size = 2
state_size = 24
seed = 123
```

```python
model = Actor(state_size, action_size, seed)
x = Variable(torch.randn(1, state_size))
y = model(x)
             
make_dot(y, params=dict(list(model.named_parameters())))
```

```python
model = Critic(state_size, action_size, seed)
x = Variable(torch.randn(1, state_size))
z = Variable(torch.randn(1, action_size))
y = model(x, z)
             
make_dot(y, params=dict(list(model.named_parameters())))
```
